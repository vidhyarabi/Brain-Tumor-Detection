```python
x_train=[]
y_train=[]
datadir='/content/drive/MyDrive/Brain_Tumor/Training'
categories=sorted(os.listdir(datadir))
```

```python
for i in categories:
  path=os.path.join(datadir,i)
  for img in os.listdir(path):
    imgpath=os.path.join(path,img)
    img_read=imread(imgpath)
    img_resize=resize(img_read,(150,150,1))
    x_train.append(img_resize)
    y_train.append(categories.index(i))
```

```python
x_test=[]
y_test=[]
datadir1='/content/drive/MyDrive/Brain_Tumor/Testing'
categories1=sorted(os.listdir(datadir1))
```

```python
for i in categories1:
  path=os.path.join(datadir1,i)
  for img in os.listdir(path):
    imgpath=os.path.join(path,img)
    img_read=imread(imgpath)
    img_resize=resize(img_read,(150,150,1))
    x_test.append(img_resize)
    y_test.append(categories1.index(i))
```

```python
x_train=np.array(x_train)
y_train=np.array(y_train)
```

```python
x_test=np.array(x_test)
y_test=np.array(y_test)
```

```python
from inspect import modulesbyfile
# Model creation
model=Sequential()
model.add(Conv2D(32,(3,3),input_shape=(150,150,1),activation='relu'))
model.add(MaxPool2D(2,2))
model.add(Flatten())
model.add(Dense(100,activation='relu'))
model.add(Dense(50,activation='relu'))
model.add(Dense(4,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=40,batch_size=50)
```

```python
model.evaluate(x_test,y_test,batch_size=20)
```

```python
img='/content/drive/MyDrive/Brain_Tumor/Testing/glioma/Te-glTr_0003.jpg'
img=imread(img)
img1=resize(img,(1,150,150,1))
y_pred=model.predict(img1)
y_pred
```
